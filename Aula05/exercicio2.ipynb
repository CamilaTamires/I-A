```python
import numpy as np # biblioteca para fazer calculos matematicos
from sklearn.linear_model import LinearRegression #biblioteca sklearn o modelo de regressao linear
import matplotlib.pyplot as plt
```

```python
# Cria um array NumPy com os valores de temperatura e o transforma em uma matriz coluna
Q = np.array([110,220,330,410,550,625,750]).reshape(-1,1)  # 'reshape(-1,1)' garante que Q tenha uma única coluna e várias linhas

# Cria um array NumPy com os valores correspondentes para a temperatura
T = np.array([30,35,40,45,60,100,150])  # Esses valores representam a variável dependente (temperatura) que queremos prever

# Instancia o objeto do modelo de regressão linear
modelo = LinearRegression()  # Cria um modelo de regressão linear

# Treina o modelo com os dados fornecidos
modelo.fit(Q, T)  # Ajusta o modelo de regressão linear aos dados de vazão (Q) e temperatura (T)

```

```python
# Testa o modelo criado

# Cria um array NumPy com novos valores de quantidade de calor para testar o modelo e o transforma em uma matriz coluna
Q_teste = np.array([100,200,300,400,500,600,700]).reshape(-1,1)  # 'reshape(-1,1)' garante que Q_teste tenha uma única coluna e várias linhas

# Faz a previsão da temperatura com base nos valores de quantidade de calor fornecidos
temperatura_prevista = modelo.predict(Q_teste)  # Usa o modelo treinado para prever a temperatura correspondente a cada valor de quantidade de calor em Q_teste

# Imprime as previsões para cada valor de quantidade de calor de teste
for i in range(len(Q_teste)):
    print(f"Quantidade de Calor: {Q_teste[i][0]} - Temperatura prevista: {temperatura_prevista[i]:.2f}")  # Mostra a quantidade de calor e a temperatura prevista correspondente

# Cria um gráfico para visualizar os dados de treinamento e as previsões
plt.scatter(Q, T, color='blue', label='Dados de treinamento')  # Plota os dados de treinamento como pontos azuis
plt.plot(Q_teste, temperatura_prevista, color='red', label='Previsões')  # Plota a linha de previsão em vermelho
plt.xlabel('Quantidade de Calor')  # Rótulo para o eixo X
plt.ylabel('Temperatura (°C)')  # Rótulo para o eixo Y
plt.title('Temperatura em Função da Quantidade de Calor')  # Título do gráfico
plt.legend()  # Adiciona uma legenda ao gráfico
plt.show()  # Exibe o gráfico
```
