```python
import numpy as np # biblioteca para fazer calculos matematicos
from sklearn.linear_model import LinearRegression #biblioteca sklearn o modelo de regressao linear
import matplotlib.pyplot as plt
```

```python
# Cria um array NumPy com os valores de renda e o transforma em uma matriz coluna
renda = np.array([1210, 5100, 8050, 10100, 30300]).reshape(-1, 1)  # 'reshape(-1, 1)' transforma o array em uma matriz coluna com uma única coluna e várias linhas

# Cria um array NumPy com os valores correspondentes para os gastos
gastos = np.array([750, 1500, 4000, 6000, 18000])  # Esses valores representam a variável dependente (gastos) que queremos prever

# Instancia o objeto do modelo de regressão linear
modelo = LinearRegression()  # Cria um modelo de regressão linear

# Treina o modelo com os dados fornecidos
modelo.fit(renda, gastos)  # Ajusta o modelo de regressão linear aos dados de renda e gastos

```

```python
# Cria um array NumPy com novos valores para testar o modelo e transforma em uma matriz coluna
renda_teste = np.array([1200, 5000, 8000, 10000, 30000]).reshape(-1, 1)  # 'reshape(-1, 1)' transforma o array em uma matriz coluna com uma única coluna e várias linhas

# Faz a previsão dos gastos com base nos novos valores de renda fornecidos
gastos_previsto = modelo.predict(renda_teste)  # Usa o modelo treinado para prever os gastos correspondentes a cada valor de renda em renda_teste

# Imprime as previsões para cada valor de renda de teste
for i in range(len(renda_teste)):
    print(f"Renda: {renda_teste[i][0]} - Gastos previstos: {gastos_previsto[i]:.2f}")  # Mostra a renda e os gastos previstos correspondentes

# Cria um gráfico para visualizar os dados de treinamento e as previsões
plt.scatter(renda, gastos, color='blue', label='Dados de treinamento')  # Plota os dados de treinamento como pontos azuis
plt.plot(renda_teste, gastos_previsto, color='red', label='Previsões')  # Plota a linha de previsão
```
