```python
import numpy as np # biblioteca para fazer calculos matematicos
from sklearn.linear_model import LinearRegression #biblioteca sklearn o modelo de regressao linear
import matplotlib.pyplot as plt
```

```python

# Cria um array NumPy com os valores de área e o transforma em uma matriz coluna
Area = np.array([55, 110, 144, 320]).reshape(-1, 1)  # 'reshape(-1, 1)' transforma o array em uma matriz coluna com uma única coluna e várias linhas

# Cria um array NumPy com os valores correspondentes para a temperatura
valor = np.array([180000, 300000, 375000, 600000])  # Esses valores representam a variável dependente (temperatura) que queremos prever

# Instancia o objeto do modelo de regressão linear
modelo = LinearRegression()  # Cria um modelo de regressão linear

# Treina o modelo com os dados fornecidos
modelo.fit(Area, valor)  # Ajusta o modelo de regressão linear aos dados de Area e valor
```

```python
# Cria um array NumPy com novos valores para testar o modelo e transforma em uma matriz coluna
Area_teste = np.array([50, 100, 120, 300]).reshape(-1, 1)  # 'reshape(-1, 1)' transforma o array em uma matriz coluna com uma única coluna e várias linhas

# Faz a previsão do valor com base nos novos valores de área fornecidos
valor_previsto = modelo.predict(Area_teste)  # Usa o modelo treinado para prever o valor correspondente a cada área em Area_teste

# Imprime as previsões para cada valor de área de teste
for i in range(len(Area_teste)):
    print(f"Área: {Area_teste[i][0]} - Valor previsto: {valor_previsto[i]:.2f}")  # Mostra a área e o valor previsto correspondente

# Cria um gráfico para visualizar os dados de treinamento e as previsões
plt.scatter(Area, valor, color='blue', label='Dados de treinamento')  # Plota os dados de treinamento como pontos azuis
plt.plot(Area_teste, valor_previsto, color='red', label='Previsões')  # Plota a linha de previsão em vermelho
plt.xlabel('Área')  # Rótulo para o eixo X
plt.ylabel('Valor')  # Rótulo para o eixo Y
plt.title('Previsões de Valor com base na Área')  # Título do gráfico
plt.legend()  # Adiciona uma legenda ao gráfico
plt.show()  # Exibe o gráfico
```
