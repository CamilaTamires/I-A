```python
import numpy as np # biblioteca para fazer calculos matematicos
from sklearn.linear_model import LinearRegression #biblioteca sklearn o modelo de regressao linear
import matplotlib.pyplot as plt
```

```python
#criando o conjunto de dados
vazao = np.array([10,20,30,5,35,40]).reshape(-1,1) # cria o conjunto de dados
nivel = np.array([50,60,70,45,80,85]) #conjunto de dados para o nivel 
#Cria modelo e treina o modelo de regressao linear
modelo = LinearRegression() #armazena a regressão linear na variavel modelo
modelo.fit(vazao, nivel) # treina o modelo de regressão linear
```

```python
#teste do modelo criado
#vazao_teste = np.array([15,25,33]).reshape(-1,1) #vazão de agua em m3/min
vazao_teste = np.array([5,2,7,30,50]).reshape(-1,1) #vazão de agua em m3/min
# fazer previsão com base na vazão de agua  
nivel_previsto = modelo.predict(vazao_teste) # realiza a previsão do nivel reservatorio

# imprimir as previsões
for i in range(len(vazao_teste)):
    print(f"vazão: {vazao_teste[i][0]} m3/min - nivel previsto: {nivel_previsto[i]}")
plt.scatter(vazao,nivel,color='blue', label='Dados de treinamento')
plt.plot(vazao_teste,nivel_previsto,color='red',label='Dados de teste')
plt.xlabel('Vazão de Agua (m3/min)')
plt.ylabel('Nivel do tanque (metros)')
plt.title('Relação entre Vazão de Agua e nivel do tanque')
```
